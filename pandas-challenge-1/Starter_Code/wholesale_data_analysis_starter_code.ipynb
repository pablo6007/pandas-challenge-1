## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
# Create a column that calculates the subtotal for each line using the unit_price and the qty
client_information_df["subtotal"] = client_information_df["unit_price"] * client_information_df["qty"]

client_information_df.head()
```

```python
# Create a column for shipping price.
# Assume a shipping price of $7 per pound for orders over 50 pounds and $10 per pound for items 50 pounds or under.


client_information_df["shipping_price"] = client_information_df["unit_weight"].apply(lambda unit_weight: unit_weight * 7 if unit_weight > 50 else unit_weight * 10)

client_information_df.head()
#print(client_information_df)
```

```python
# Create a column for the total price using the subtotal and the shipping price along with a sales tax of 9.25%
sales_tax = 9.25 / 100


client_information_df["sales_tax"] = client_information_df["subtotal"] * sales_tax

client_information_df["total_price"] = client_information_df["subtotal"] + client_information_df["shipping_price"]  + client_information_df["sales_tax"]

client_information_df.head()
```

```python
# Create a column for the cost of each line using unit cost, qty, and
# shipping price (assume the shipping cost is exactly what is charged to the client).

client_information_df["line_cost"] = client_information_df["unit_cost"] * client_information_df["qty"] * client_information_df["shipping_price"]

client_information_df.head()
```

```python
# Create a column for the profit of each line using line cost and line price

client_information_df["line_profit"] = client_information_df["line_cost"] * client_information_df["total_price"]

client_information_df.head()
```
